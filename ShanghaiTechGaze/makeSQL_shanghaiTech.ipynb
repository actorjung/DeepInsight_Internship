# SQL작성

```python
import sys
import os
import pandas as pd
import cv2
from glob import glob
import EDA_utils as EU

# query 실행
query1 = (f'''CREATE TABLE shanghaiTechGaze_info (
        id INT(11) NOT NULL AUTO_INCREMENT primary key,
        img_dir VARCHAR(100) NOT NULL,
        img_format VARCHAR(10) NOT NULL,
        img_width INT(11) NOT NULL,
        img_height INT(11) NOT NULL,
        color_space VARCHAR(10) NOT NULL,
          
        label_dir VARCHAR(100) NOT NULL,
        label_format VARCHAR(10) NOT NULL,
        WhichEye VARCHAR(10) NOT NULL,
        x_gt VARCHAR(1000) NOT NULL,
        y_gt VARCHAR(1000) NOT NULL)''')
try:
    EU.MakeSQL.create_table(query1)
except:
    pass

query2 = (f'''INSERT INTO shanghaiTechGaze_info 
        (img_dir, img_format, img_width, img_height, color_space,
          label_dir, label_format, WhichEye, x_gt, y_gt) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)''')


# 나는 라벨 파일이 없으므로 각각 다른경로에서 불러서 데이터를 저장해야 함 텍스트를 만들어서 그걸 불러오면 형식이 통일되니까 그렇게 하는게 나을듯?
# 먼저 라벨 파일을 만드는게 1번째 그 라벨 파일을 에스큐엘에다 저장해야겠다 
# 라벨파일 

label_file = "shanghaiTech_label.csv"
# print(label_file)

value_list1 = []
error_list = []

try:
    with open(label_file, 'r') as file:
        texts = file.readlines()
    for line in texts[1:]:
        image_path, image_format, width, height, color_area, eye_position, label_path, label_format, x_gt, y_gt = line.split(',')
        img_dir = image_path
        img_format = image_format
        img_width = width
        img_height = height
        color_space = color_area
        label_dir = label_path.replace("nsmount/shanghaiTechGaze/data/images/coordinate","nsmount/shanghaiTechGaze/data/RGB/labeldata")
        WhichEye = eye_position
        x_gt = x_gt.replace('"',"")
        y_gt = y_gt.replace('"',"")
        label_format = "mat"

        value = (img_dir, img_format, img_width, img_height, color_space,  label_dir, label_format, WhichEye, x_gt, y_gt)
        # print(value)
        value_list1.append(value)

        # query 실행

        if len(value_list1) >= 500:
            print(value_list1[0])
            EU.MakeSQL.insert_dataset_values(query2, value_list1)
            value_list1 = []
            print(f'{len(value_list1)} / {len(label_file)}')
            
except FileNotFoundError:
    print(f"{label_file} 파일을 찾을 수 없습니다.")
except Exception as e:
    print(f"오류 발생: {e}")
    
EU.MakeSQL.insert_dataset_values(query2, value_list1)

# print(value_list2)
value_list1 = []
print('done')
```
